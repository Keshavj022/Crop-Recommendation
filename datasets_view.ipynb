```python
output_csv_path = "State District.csv"
```

```python
# Extract unique state names and their corresponding districts
state_districts = landuse.groupby("State Name")["Dist Name"].unique().to_dict()

# Convert arrays to lists for better readability
state_districts = {state: list(districts) for state, districts in state_districts.items()}

# Display a preview of the extracted data
state_districts
```

```python
import csv
with open(output_csv_path, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["State", "District"])  # Writing the header row

    for state, districts in state_districts.items():
        for district in districts:
            writer.writerow([state, district])  # Writing each district in a separate row

print(f"CSV file saved as {output_csv_path}")
```
